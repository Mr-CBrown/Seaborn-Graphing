```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
sns.set(color_codes=True) #set color set in seaborn
```

```python
sales = pd.read_csv("sales_data_sample.csv", sep=",", encoding='Latin-1') 
#create data frame and import data set from kaggle.. modified encoding from UTF-8 due to error
```

```python
sales.head() # see top rows of imported data
```

```python
sales.info() #View data type and column names of dataset
```

```python
ax= sns.barplot(x ='PRODUCTLINE',y = 'SALES', data = sales, palette = 'plasma')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")#univariate function
#Barchart of the quarterly sales 
```

```python
ax = sns.histplot(sales['PRICEEACH'])  # Histogram of sales price per units sold
```

```python
 sns.jointplot(data = sales, x='PRICEEACH',y='SALES') #Bivariate Function
```

```python
sns.jointplot(data = sales, x='PRICEEACH',y='SALES', kind = 'kde') #Same graph... Changing the graph kind to KDE
```

```python
sns.pairplot(sales, hue="PRODUCTLINE")
```

```python
sns.stripplot(data = sales, y='PRODUCTLINE', x='SALES')
```

```python
sns.boxplot(data = sales, y='PRODUCTLINE', x='PRICEEACH')
```

```python
ax = sns.countplot(data =sales, x='PRODUCTLINE')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right") #rotate the x axis to make it readable
```
